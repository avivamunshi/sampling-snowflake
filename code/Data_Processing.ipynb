# Data Preprocessing

Formatting the data in the columns so they don't give us trouble when we load it into Snowflake.

```python
import pandas as pd
```

```python
data = pd.read_excel("/Users/aviva/Desktop/MSDS/quarter_4/Scalable Data Systems/Project/final_debt_data.xlsx")
data.head()
```

```python
data['Debt'] = data.loc[:, 1960:2022].sum(axis=1)
```

```python
df = data.groupby(['Country Code','Country Name', 'Indicator Code','Indicator Name'])['Debt'].sum().reset_index()
```

```python
df.head()
```

```python
df['Indicator Name'] = df['Indicator Name'].str.replace(',', ';')
df['Country Name'] = df['Country Name'].str.replace(',', '')
```

```python
df["Indicator Name"].nunique()
```

```python
df.head()
```

```python
df.to_csv('/Users/aviva/Desktop/MSDS/quarter_4/Scalable Data Systems/Project/snowflake_input_data.csv', index=False)
```

```python
max_length = 0
```

```python
for index, row in df.iterrows():
    # Join all columns in the row to create a single string
    row_string = ''.join(map(str, row))
    
    # Calculate the length of the concatenated string
    row_length = len(row_string)
    
    # Update max_length if the current row has a greater length
    if row_length > max_length:
        max_length = row_length

print("Max number of characters in a row across all columns:", max_length)
```
